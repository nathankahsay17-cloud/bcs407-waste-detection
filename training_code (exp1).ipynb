```python
import torch, random, numpy as np, os

seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
os.environ["PYTHONHASHSEED"] = str(seed)

print("Seed set to", seed)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")

results_exp1 = model.train(
    data="/content/waste_detection/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    seed=42,
    project="/content/runs_project2",
    name="exp1_yolov8m"
)
```

```python
model = YOLO("/content/runs_project2/exp1_yolov8m/weights/best.pt")
metrics_exp1 = model.val(
    data="/content/waste_detection/data.yaml",
    project="/content/runs_project2",
    name="exp1_yolov8m_val"
)
print(metrics_exp1.results_dict)
```
