```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

results_exp2 = model.train(
    data="/content/waste_detection/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    seed=42,
    project="/content/runs_project2",
    name="exp2_more_paper_cardboard"
)
```

```python
from ultralytics import YOLO

model = YOLO("/content/runs_project2/exp2_more_paper_cardboard/weights/best.pt")
metrics_exp2 = model.val(
    data="/content/waste_detection/data.yaml",
    project="/content/runs_project2",
    name="exp2_more_paper_cardboard_val"
)

print(metrics_exp2.results_dict)
```

```python
from ultralytics import YOLO

model = YOLO("yolov8m.pt")

results_final = model.train(
    data="/content/waste_detection/data.yaml",
    epochs=50,
    imgsz=640,
    batch=8,
    workers=2,
    cache=False,
    seed=42,
    project="/content/runs_project2",
    name="final_more_paper_cardboard"
)
```
